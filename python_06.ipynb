# **그룹화 (Grouping)**

```
df.head()
```

```
df.shape
```

```
df=df.rename(columns={'분양가격(㎡)':'분양가'})
```

```
df.dtypes
```

```
df['분양가'] = df['분양가'].apply(pd.to_numeric, errors = 'coerce')
#찾아서 다시 해보기
```

```
df.groupby(['지역명', '연도', '월'])['분양가'].agg('sum')
```

```
df.info()
```

```
grouped = df.groupby(['지역명', '연도', '월']).sum()
grouped
```

```
grouped.head()
```

```
cc=grouped.stack()
cc.head(20)
#위에서 밑으로 쌓는 함수
```

```
grouped.unstack(0)
#왼쪽에서 오른쪽으로 쌓는다
#()안에는 column 을 정해준다
```

group이 되어 있는 DataFrame을 unstack의 ‘레벨’을 조정하여 보여줄 수 있다. 위의 grouped에서 ‘도시’ = level 1, ‘연도’ = level 2, ‘월’

```
# level1: 열을 연도별로 출력
grouped.unstack(1)
```

```
# level2: 열을 연도별로 출력
grouped.unstack(2)
```

# **Pivot Table**

```
df.head(10)
```

```
pd.pivot_table(df, values='분양가', index=['연도','지역명'], columns=['월'])
```

```
df.head(10)
```

```
df['분양가'].describe()
```

```
df['평가'] = 0
```

```
df.head(10)
```

가격대 별 **평가**

low: 25% = 2387 보다 싼 분양가

mid: 50% = 2387 ~ 3130

high: 75% = 3130 ~ 3383

very high: 75% ~ 100% = 3383 보다 비싼 분양

```
conditions = [
    (df['분양가'] < 2387),
    (df['분양가'] >= 2387) & (df['분양가'] < 3130),
    (df['분양가'] >= 3130) & (df['분양가'] < 3383),
    (df['분양가'] >= 3383),
    (df['분양가'] == np.nan)
]
choices = ['저렴', '보통', '비쌈', '매우 비쌈', '-']
df['평가'] = np.select(conditions, choices, default=0)
```

```
df.head(20)
```

```
df.groupby(by='평가').count()
```

```
df.sort_values(by='분양가', ascending=False)[:10]
#내림가 
```

```
df.dtypes
```
